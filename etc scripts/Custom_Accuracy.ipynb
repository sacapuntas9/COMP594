```python
import cv2
import numpy as np
import os
import sys
from skimage import measure


directorySeparator = '\\'

generatedMaskDir = "E:\\594_data\\\smoothmuscle_NN_Data\\test_holdout\\classified" #"E:\\594_data\\combined_ground_truth\\automated_masks"
groundTruthDir = "E:\\594_data\\\smoothmuscle_NN_Data\\test_holdout\\masks"



directoryMask = os.fsencode(generatedMaskDir)
directoryTruth = os.fsencode(groundTruthDir)
```

```python
def giveNumberFirstInSecond( first,second ):
    #expects labelsTruth or labelsMask as first parameter, and imageMask or ImageTruth as second parameter.
    #checks if some pixel in each of the labels exists in the mask; returns the count of how many do (hits) 
                                                                    #and the count of how many do not (misses)
    hits = 0
    misses = 0
    
    for label in np.unique(first):
            # if this is the background label, ignore it
            if label == 0:
                continue
            
            # otherwise, construct the label mask 
            
            labelMask = np.zeros(second.shape, dtype="uint8")
            #create another blank array with the shape of the image
            
            labelMask[first == label] = 255
            #make anywhere where label is the label in this pass is equal in the label list = 255 in the new mask
            
            nonzeroes = np.transpose(np.nonzero(labelMask))
            #create an array with the format [[x,y][x,y]...] for each pixel in the current connected component
            
            
            
            noHit = True
            
            for el in nonzeroes: #iterate over the pixels in this connected component and check them against the comparison mask
                    x = el[0]
                    y = el[1]
                    if second[x][y] == 255:
                        hits = hits + 1
                        noHit = False
                        break
            
            if noHit:
                misses = misses + 1
                    
    return (hits,misses)
                        
```

```python
f1Total=0
fileCounter=0

for filename in os.listdir(directoryMask):
    file = os.fsdecode(filename)
    if file.endswith(".tif"): 
        imageMask = cv2.imread(generatedMaskDir+ directorySeparator + file, -1)
        imageTruth = cv2.imread(groundTruthDir+ directorySeparator + file, -1)
        
        if imageTruth is None:
            print(filename)

        labelsTruth = measure.label(imageTruth, neighbors=8, background=0)
        labelsMask = measure.label(imageMask, neighbors=8, background=0)
        
        numTruth = len(np.unique(labelsTruth)) - 1
        numMask = len(np.unique(labelsMask)) - 1
        
        
        #print(str(numTruth) + " True items and " + str(numMask) + " Generated Items.")
        
        inMaskInGT, inMaskNotInGT = giveNumberFirstInSecond(labelsMask,imageTruth)
        
        inGTInMAsk, inGTNotInMask = giveNumberFirstInSecond(labelsTruth, imageMask)
        

        truePositives = inMaskInGT
        falsePositives = inMaskNotInGT
        falseNegatives = inGTNotInMask
        
        if (truePositives + falsePositives == 0):
            precision = 0
        else:
            precision = truePositives / (truePositives + falsePositives)
        
        if (truePositives + falseNegatives == 0):
            recall = 0
        else:
            recall = truePositives / (truePositives + falseNegatives)
        

        if precision + recall == 0:
            F1 = 0
        else:
            F1 = 2*precision*recall / (precision + recall)
        
        #print(F1)
        
        f1Total += F1
        fileCounter += 1
        
f1_score = f1Total / fileCounter
print("F-Score: "+str(f1_score)) 
```
